```python
import pandas as pd
import matplotlib as plt
from matplotlib import pyplot as plt
import datetime as dt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('covid_19_india.csv', parse_dates=['Date'], dayfirst=True)
```

```python
df.head()
```

```python
df = df[['Date','State/UnionTerritory','Cured','Deaths','Confirmed']]
```

```python
df.columns= ['date','state','cured','deaths','confirmed']
```

```python
df.head()
```

```python
today = df[df.date== '2020-03-17']
```

```python
today
```

```python
cases = today.sort_values(by="confirmed",ascending=False)
```

```python
cases
```

```python
top_states= cases[0:5]
```

```python
top_states
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x="state",y="confirmed",data=top_states,hue="state")
```

```python
maha = df[df.state=='Maharashtra']
```

```python
maha
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="date",y="deaths",data=maha,color="r")
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="date",y="confirmed",data=maha,color="g")
```
